```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "C:\\Users\\furqa\\Downloads\\company_sales_data.csv"
data = pd.read_csv(path)
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
#1.Exercise 1: Read Total profit of all months and show it using a line plot
df_profit = df[['month_number', 'total_profit']]
df_profit.set_index('month_number', inplace=True)

df_profit.plot(kind='line')

plt.title('Month vs Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.show()
```

```python
#Exercise 2: Get total profit of all months and show line plot with the following Style properties:

df_profit.plot(kind='line', linestyle='dotted', 
               color='red', marker ='o',
               markerfacecolor='red',linewidth= 1.5)

plt.title('Month vs Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.legend(loc='lower right') 

plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot
df_units_sold = df[['month_number', 'facecream', 'facewash',
                    'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
df_units_sold.set_index('month_number', inplace=True)
df_units_sold

df_units_sold.plot(kind='line', figsize=(10, 6))

plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(df['month_number'])

plt.show()
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
x = df[['month_number']]
y = df[['toothpaste']]

df.plot(kind='scatter', x='month_number', y='toothpaste', 
        label = 'Toothpaste Sales')

plt.title('Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of Toothpaste Sold')
plt.legend(loc='upper left')
plt.xticks(df['month_number'])

plt.grid(True, linewidth= 1, linestyle="--")

plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
df_fcfw = df[['month_number', 'facecream', 'facewash']]
df_fcfw.set_index('month_number', inplace=True)

df_fcfw.plot(kind='bar', figsize=(10, 6), rot=0)

plt.title('Sales Data of Face Cream & Face Wash')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(df['month_number'])

plt.show()
```

```python
#Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
df_bs = df[['month_number', 'bathingsoap']]
df_bs.set_index('month_number', inplace=True)

df_bs.plot(kind='bar', figsize=(10, 6), rot=0)

plt.title('Sales Data of Bathing Soap')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(df['month_number'])

plt.savefig('C:\\Users\\Muham\\Downloads\\sales_data_bathingsoap.png', dpi=300)
plt.show()
```

```python
#Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
df_profit = df[['month_number', 'total_profit']]
df_profit.set_index('month_number', inplace=True)

df_profit.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Each Month Profit')
plt.xlabel('Total Profit')

plt.show()
```

```python
#Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
df_total_sales = df[['facecream', 'facewash', 'toothpaste',
                     'bathingsoap', 'shampoo', 'moisturizer']].sum()

df_total_sales.plot(kind='pie',
                    figsize=(5, 6),
                    autopct='%1.1f%%',
                    startangle=90,
                   )

plt.title('Total Sales Data For Each Product')
plt.axis('off')

plt.show()
```

```python
#Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

f, ax = plt.subplots(2, sharex=True)
ax[0].plot(df[['month_number']], df[['bathingsoap']],
           label = 'Bathing Soap Sales', color='red')
ax[1].plot(df[['month_number']], df[['facewash']],
           label = 'Face Wash Sales', color='blue')

ax[0].set_title('Sales data of Bathing Soap')
ax[0].set_ylabel('Sales Units')
ax[0].set_xticks(df['month_number'])

ax[1].set_title('Sales data of Face Wash')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Sales Units')
ax[1].set_xticks(df['month_number'])

plt.show()
```

```python
#Exercise Question 10: Read all product sales data and show it using the stack plot
df_units_sold = df[['month_number', 'facecream', 'facewash',
                    'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
df_units_sold.set_index('month_number', inplace=True)
df_units_sold

df_units_sold.plot(kind='area',
                   stacked=True,
                   figsize=(20, 10))

plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.xticks(df['month_number'])

plt.show()
```
