```python
import pandas as pd
import sqlite3
```

```python
# Checking to make sure connection is properly established
try:
    sqliteconnect = sqlite3.connect('Hospitalinfo_2.db')
    cursor = sqliteconnect.cursor()
    print("Connection successful.")
    sqlite_select_query = 'select sqlite_version();'
    cursor.execute(sqlite_select_query)
    record = cursor.fetchall()
    print("Database version:", record)
    
except sqlite3.Error as error:
    print("Error occurred while connecting to sqlite", error)
```

```python
# Direct connection
con = sqlite3.connect("Hospitalinfo_2.db")
```

```python
# Read Doctor Table from database
doctors = pd.read_sql_query("SELECT * FROM Doctor", con)
doctors
```

```python
# Read Hospital Table from database

hospital = pd.read_sql_query("SELECT * FROM Hospital", con)
hospital
```

```python
# Get all information on the hospitals

query = "SELECT * FROM Hospital \
        JOIN Doctor \
        ON Hospital.ID = Doctor.HospitalID"

ex_2 = pd.read_sql_query(query, con)
ex_2
```

```python
# List doctor by speciality and salary

query = "SELECT DocName,Speciality,Salary \
        FROM Doctor \
        ORDER BY Speciality"

ex_3 = pd.read_sql_query(query, con)
ex_3
```

```python
# List doctor by speciality and salary

info1 = input("What is their speciality? ")
info2 = input("What is their salary? ")

query = f"SELECT DocName FROM Doctor \
        WHERE Speciality = '{info1}' \
            AND salary = {info2}"

ex_3_1 = pd.read_sql_query(query, con)
ex_3_1
```

```python
# Get all the doctors by hospital

hospital_name = input("What is the hospital name? ")
query = f"SELECT d.DocName, d.HospitalID, h.Name \
        FROM Doctor d \
        JOIN Hospital h \
        ON d.HospitalID = h.ID \
        WHERE h.Name = '{hospital_name}'"

ex_4 = pd.read_sql_query(query, con)
ex_4
```
